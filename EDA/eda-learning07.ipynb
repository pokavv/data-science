```python
%matplotlib inline
import pandas as pd

df = pd.read_csv('../data/laptops.csv')
```

```python
brand_nation = {
    'Dell':'U.S.',
    'Apple':'U.S.',
    'Acer':'Taiwan',
    'HP':'U.S.',
    'Lenovo':'China',
    'Alienware':'U.S.',
    'Microsoft':'U.S.',
    'Asus':'Taiwan'
}
```

```python
df['brand'].map(brand_nation) # map(x): x를 기준으로 map
```

```python
df['brand_nation'] = df['brand'].map(brand_nation)
df
```

```python
nation_groups = df.groupby('brand_nation')
type(nation_groups) # nation_groups의 type
```

```python
nation_groups.count() # count
```
