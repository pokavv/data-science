# scatter plot (산점도)

* showing specific correlations
* for example, when distributed in a straight line, there is a strong association, and as it is scattered, the association becomes weaker

```python
%matplotlib inline
import pandas as pd

df = pd.read_csv('../data/exam.csv')
df.head()
```

```python
df.plot(kind='scatter', x='math score', y='writing score') # 서로 연관성을 확인하기 위한 x, y축 설정
```

```python
df.plot(kind='scatter', x='reading score', y='writing score')
```
