```python
import pandas as pd
import seaborn as sns

df = pd.read_csv('../data/laptops.csv')
df.head()
```

```python
df['os'].unique()
```

```python
sns.catplot(data=df, x='os', y='price', kind='box') # catplot: categorical plot (category 별 visualization)
```

```python
sns.catplot(data=df, x='os', y='price', kind='strip') # strip plot: 점 분포 plot
```

```python
df['processor_brand'].unique()
```

```python
sns.catplot(data=df, x='os', y='price', kind='strip', hue='processor_brand') # hue: color 의미
```

```python
sns.catplot(data=df, x='os', y='price', kind='swarm', hue='processor_brand') # swarm plot: 겹치지않고 퍼뜨림
```
