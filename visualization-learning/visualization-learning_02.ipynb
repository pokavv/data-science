# bar graph

* to compare the size of values

```python
%matplotlib inline
import pandas as pd
```

```python
df = pd.read_csv('../data/sports.csv', index_col = 0)
df
```

```python
df.plot(kind = 'bar')
```

```python
df.plot(kind = 'barh') # 'barh' = bar horizontal graph
```

```python
df.plot(kind = 'bar', stacked = True) # 한 개의 bar에 stack
```

```python
df['Female'].plot(kind = 'bar') # 특정 group 만을 출력 (Serise)
```
