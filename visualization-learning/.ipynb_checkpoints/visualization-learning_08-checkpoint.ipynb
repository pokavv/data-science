```python
%matplotlib inline
import pandas as pd
import seaborn as sns

df = pd.read_csv('../data/body.csv', index_col=0)
df.head()
```

```python
df.plot(kind='hist', y='Height', bins=15)
```

```python
sns.distplot(df['Height'], bins=15) # distribution plot
```

```python
df.plot(kind='box', y='Height')
```

```python
sns.violinplot(y=df['Height']) # violinplot
```

```python
df.plot(kind='scatter', x='Height', y='Weight')
```

```python
sns.kdeplot(df['Height'], df['Weight']) # kdeplot 등고선
```
