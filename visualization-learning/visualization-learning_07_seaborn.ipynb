# Seaborn (Statistical Data Visualization, 통계 기반 데이터 시각화)

## PDF (Probability Density Function, 확률 밀도 함수)
* probability density function represents the distribution of the dataset.
* probability of a particular section is equal to the area of that section under the graph.
* sum of all areas under the graph equals 1.
* since the probability that a single estimate presented as a point estimate agrees with the parameter is almost zero, the probability that a particular value exists rather than the probability for a particular range is zero.

## KDE (Kernel Density Estimation, 커널밀도추정)
* one of the non-parametric density estimation methods, and it is a method to improve the problem of the histogram method using a kernel function.
* kernel function: mathematically, defined as a non-negative (always positive) function that is symmetric about the origin and has an integral value of 1.

```python
import pandas as pd
import seaborn as sns

df = pd.read_csv('../data/body.csv', index_col=0)
df.head()
```

```python
df['Height'].value_counts().sort_index() # sort_index() : sort index
```

```python
sns.kdeplot(df['Height'], bw=0.5) # bw : bandwidth (대역폭)
```
