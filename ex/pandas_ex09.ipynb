```python
import pandas as pd
df = pd.read_csv('../data/enrolment_1.csv')
df
```

```python
df['status'] = 'allowed' # 'status' column 추가, default value = 'allowed'
df.loc[(df['year'] == 1) & (df['course name'] == 'information technology'), 'status'] = 'not allowed'
df.loc[(df['year'] == 4) & (df['course name'] == 'commerce'), 'status'] = 'not allowed'
```

```python
allowed = df['status'] == 'allowed' # status가 allowed인 data만 모으기
count = df.loc[allowed, 'course name'].value_counts() # value_count()로 각 과목별 신청인원 확인
not_allowed_course = list(count[count < 5].index) # 과목별 신청인원 5 이하인 과목의 index 추출하여 list 생성

for course in not_allowed_course: # for 문을 통해 not allowed 삽입
    df.loc[df['course name'] == course, 'status'] = 'not allowed'
```
